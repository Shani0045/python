```python
from tkinter import *
from PIL import ImageTk, Image
import cv2
root = Tk()
root.geometry("600x500")
# Create a frame
app = Frame(root, bg="white",relief=RIDGE,bd=10)
app.place(x=150,y=10,width=300,height=300)
# Create a label in the frame
def cancel():
    lmain.place_forget()
    
lmain = Label(app)
lmain.place(x=0,y=0)
# Capture from camera
cap = cv2.VideoCapture(0)
print(cap)
# function for video streaming
c=0
def video_stream():
    global c
    _, frame = cap.read()
    cv2image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    img = Image.fromarray(cv2image)
    imgtk = ImageTk.PhotoImage(image=img.resize((277,277)))
    lmain.imgtk = imgtk
    lmain.configure(image=imgtk)
    cv2.imwrite(f"shani/{c}-img.png",cv2image)
    a=lmain.after(10, video_stream)
    if c==50:
        root.after_cancel(a)
    c+=1
btn=Button(root,text="Start",font=("ariel 15 bold"),command=video_stream)
btn.place(x=200,y=350)
btn=Button(root,text="cancel",font=("ariel 15 bold"),command=cancel)
btn.place(x=300,y=350)

root.mainloop()
```

```python
import tkinter

root = tkinter.Tk()
root.title('Timer')
root.state('zoomed')

sec = 0
doTick = True

def tick():
    global sec
    if not doTick:
        return
    sec += 0.1
    sec = round(sec,1)
    timeLabel.configure(text=sec)
    root.after(100, tick)

def stop():
    global doTick
    doTick = False

def start():
    global doTick
    doTick = True
    # Perhaps reset `sec` too?
    tick()

timeLabel = tkinter.Label(root, fg='green',font=('Helvetica',150))
timeLabel.pack()

startButton = tkinter.Button(root, text='Start', command=start)
startButton.pack()

stopButton = tkinter.Button(root, text='Stop', command=stop)
stopButton.pack()

root.mainloop()
```
