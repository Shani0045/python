```python
from tkinter import*
from tkinter import ttk
root=Tk()
root.geometry("600x500")

style=ttk.Style()
style.configure("TLabel",foreground="blue",font=("ariel",15,"bold"))
l=ttk.Label(root,text="Username")
l.grid(row=0,column=0)

e=ttk.Entry(root)
e.grid(row=0,column=1,padx=5)

l=ttk.Label(root,text="Password")
l.grid(row=1,column=0)

e=ttk.Entry(root)
e.grid(row=1,column=1,padx=5)
root.mainloop()
```

```python
from tkinter import*
from tkinter import ttk
root=Tk()
root.geometry("600x500")

style=ttk.Style()
print(style.theme_names())
# style.theme_use("xpnative")
style.configure("new.TLabel",foreground="blue",font=("ariel",15,"bold"))
style.configure("TEntry",foreground="blue",font=("ariel",15,"bold"))

l=ttk.Label(root,text="Username")
l.grid(row=0,column=0)

e=ttk.Entry(root)
e.grid(row=0,column=1,padx=5)
l=ttk.Label(root,text="Password",style="new.TLabel")
l.grid(row=1,column=0)

e=ttk.Entry(root)
e.grid(row=1,column=1,padx=5)
root.mainloop()
```

```python
from tkinter import*
from tkinter import ttk
root=Tk()
root.geometry("600x500")

style=ttk.Style()
style.configure("TLabel",foreground="blue",font=("ariel",15,"bold"))
style.configure("TEntry",foreground="blue",font=("ariel",15,"bold"))

l=ttk.Label(root,text="Username")
l.grid(row=0,column=0)

bullet="\u2022"
e=ttk.Entry(root,show=bullet)
e.grid(row=0,column=1,padx=5)


root.mainloop()
```
