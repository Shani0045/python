# linear algebra1

```python
from scipy import linalg
```

```python
import numpy as np
```

## 5x=10
what is value of x

```python
a=np.array([5])
b=np.array([10])
linalg.solve(a,b)
```

```python
np.linalg.solve([a],b)
```

## 3x+4y=79
2x+9y=24
what is the value of x and y

```python
x=np.array([[3,4],[2,9]])
y=np.array([79,24])
linalg.solve(x,y)
```

```python
d=np.linalg.solve(x,y)
print("x =",d[0],"y =",d[1])
```

# eigen value

```python
a=linalg.eig(y) # return eigen value and eigen vector
print("eigen value ",a[0])
print("\neigen vector\n\n",a[1])
```

```python
linalg.eigvals(y)  # eigen value
```
