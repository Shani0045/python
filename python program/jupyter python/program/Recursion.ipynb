```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)

factorial(5)
```

# add all natural number between 1 to 10

```python
def number(n):
    if n==1:
        return 1
    return n+number(n-1)
number(10)
```
