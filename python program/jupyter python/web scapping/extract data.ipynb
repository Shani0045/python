```python
import requests as r
from bs4 import BeautifulSoup
```

```python
url=r.get("https://en.wikipedia.org/wiki/List_of_national_capitals")
```

```python
soup=BeautifulSoup(url.text,"html.parser")
```

```python
print(soup.prettify())
```

```python
print(soup.title)
```

```python
for link in soup.findAll("a"):
    print(link.get("href"))
```

```python
soup.get_text()
```

```python
import re
```

```python
table=soup.find_all("table")[1]
```

```python
rows=table.find_all("tr")
```

```python
rows
```

```python
rows_list=list()
```

## iterate through all the rows in table

```python
for tr in rows:
    td=tr.find_all("td")
    row=[i.text for i in td]
    rows_list.append(row)
```

```python
rows_list
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(rows_list,columns=["city","country","notes"])
```

```python
df.set_index("country",inplace=True)
```

```python
df
```

```python
df.to_csv("beutyful.csv")
```
