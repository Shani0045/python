```python
from bs4 import BeautifulSoup
import requests as r
```

```python
web=r.get("https://www.flipkart.com/search?q=mi%20mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
```

```python
soup=BeautifulSoup(web.text,"html.parser")
```

```python
soup.prettify()
```

```python
for i in soup.find_all("div",{"class":"_4rR01T"}):
    print(i.text)
```

```python
div_class=soup.find_all('div',{"class":"_4rR01T"})   # is class me 
for i in div_class:
    print(i.get_text())
```

```python
mobile_phone=[]

div_class=soup.find_all('div',{"class":"_4rR01T"})   # is class me 
for i in div_class:
    print(i.text)
    mobile_phone.append(i.text)
```

```python
mobile_phone
```

```python
for i in soup.find_all('div',{"class":"_30jeq3 _1_WHN1"}):
    a=i.text.replace("₹","")
    print(a)
          
```

```python
mobile_price=[]
div_class=soup.find_all('div',{"class":"_30jeq3 _1_WHN1"})   # is class me 
for i in div_class:
    print(i.text)
    mobile_price.append(i.text.replace("₹",""))
```
