## scrapp samsung phones

```python
rsp=r.get("https://www.flipkart.com/search?q=samsung%20phone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
rsp
```

```python
soup=BeautifulSoup(rsp.content,"html.parser")
```

```python
print(soup.prettify())
```

```python
phones=soup.select("._4rR01T")
phones
```

```python
for i in phones:
    print(i.text)
```

```python
phone_name=map(lambda x:x.text,phones)
phon=list(phone_name)
```

```python
phon
```

```python
prices=soup.select("._30jeq3")
prices
```

```python
price=map(lambda x:x.text,prices)
price=list(price)
```

```python
price
```

```python
phon
```
