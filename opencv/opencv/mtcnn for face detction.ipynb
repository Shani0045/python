```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle,Circle
from mtcnn.mtcnn import MTCNN
```

```python
import mtcnn
mtcnn.__version__
```

```python
def draw_image(file_name,result_list):
    data=plt.imread(file_name)
    plt.imshow(data)
    ax=plt.gca()
    for result in result_list:
        x,y,w,h=result["box"]
        rect=Rectangle((x,y),w,h,fill=False,color='red')
        ax.add_patch(rect)
    plt.show()
    
file="f:/varun1.jpg"
img=plt.imread(file)
detector=MTCNN()
faces=detector.detect_faces(img)
draw_image(file,faces)
```

# using opencv

```python
from mtcnn.mtcnn import MTCNN
import cv2
detector=MTCNN()
```

```python
img=cv2.imread("f:/varun1.jpg")
results=detector.detect_faces(img)
results
```

```python
bounding_box=results[0]["box"]
bounding_box
```

```python
keypoints=results[0]["keypoints"]
keypoints
```

```python
cv2.rectangle(img,(bounding_box[0],bounding_box[1]),(bounding_box[0]+bounding_box[2],bounding_box[1]+bounding_box[3]),(0,255,0),5)
cv2.imshow(" ",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# video camera

```python
from mtcnn.mtcnn import MTCNN
import cv2
detector=MTCNN()
```

```python
cap=cv2.VideoCapture(0)
while(True):
    _,frame=cap.read()
    faces=detector.detect_faces(frame)
    bounding_box=faces[0]["box"]
#     for key,value in faces[0]['keypoints'].items():
    cv2.rectangle(frame,(bounding_box[0],bounding_box[1]),(bounding_box[0]+bounding_box[2],bounding_box[1]+bounding_box[3]),(0,255,0),5)
#         cv2.circle(frame,value,6,(0,255,0),5)
    cv2.imshow(" ",frame)
    if cv2.waitKey(1)==ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```
