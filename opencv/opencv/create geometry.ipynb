```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=np.zeros((400,400,3),dtype="uint8")
cv2.copyTo()
```

```python
plt.imshow(img)
```

```python
a=cv2.line(img,(0,0),(100,100),(0,255,0),5)
plt.imshow(a)
```

```python
cv2.rectangle(img,(25,25),(250,250),(225,0,0),3)
plt.imshow(img)
```

```python
cv2.circle(img,(100,100),50,(255,0,0),3)
plt.imshow(img)
```

```python
cv2.putText(img,"shani",(50,150),cv2.FONT_HERSHEY_SIMPLEX,3,(255,0,0),3)
plt.imshow(img)
```

```python
a=cv2.rectangle(img,(50,100),(300,155),(225,0,0),3)
cv2.putText(img,"shani",(55,150),cv2.FONT_HERSHEY_SIMPLEX,3,(255,0,0),3)
plt.imshow(a)
```
